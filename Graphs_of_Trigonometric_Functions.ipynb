# Graphs of Trigonometric Functions with Python

In this tutorial, you will learn how to plot the graphs of the basic trigonometric functions—**sine**, **cosine**, and **tangent**—using Python. We'll use the `matplotlib` and `numpy` libraries to visualize these functions and explore their key properties, including **amplitude**, **period**, and **phase**.

Import the necessary modules:

````python
import numpy as np
import matplotlib.pyplot as plt
````

## 2. Plotting the Sine, Cosine, and Tangent Functions

Let's plot all three functions on the same graph over the interval from 0 to $2\pi $ radians (0° to 360°).

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 values from 0 to 2π
x = np.linspace(0, 2 * np.pi, 1000)

# Calculate sine, cosine, and tangent values
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')
plt.plot(x, y_tan, label='tan(x)', alpha=0.5)  # Make tangent a bit transparent

# Limit y-axis for better visualization of tan(x)
plt.ylim(-2, 2)

plt.title("Graphs of sin(x), cos(x), and tan(x)")
plt.xlabel("x (radians)")
plt.ylabel("Function value")
plt.legend()
plt.grid(True)
plt.show()
````

## 3. Understanding Periodic Behavior

### Sine and Cosine

- **Period**: Both `sin(x)` and `cos(x)` repeat every $2\pi$ radians (360°).
- **Amplitude**: Both have a maximum value of 1 and a minimum of -1.
- **Phase Shift**: The cosine graph is the same as the sine graph, but shifted left by $\pi/2$ radians (90°).

### Tangent

- **Period**: The tangent function repeats every $\pi$ radians (180°).
- **Asymptotes**: The graph of `tan(x)` has vertical asymptotes where the function is undefined (at $x = \frac{\pi}{2}, \frac{3\pi}{2}, ...$).
- **No amplitude**: Tangent's range is all real numbers.


## 4. Plotting Sine and Cosine Together

````python
x = np.linspace(0, 2 * np.pi, 1000)
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.title('Sine vs Cosine')
plt.xlabel('x (radians)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
````

## 5. Plotting Tangent with Asymptotes

It's helpful to show where tangent is undefined. Let's plot vertical lines at $x = \frac{\pi}{2}$ and $x = \frac{3\pi}{2}$:

````python
x = np.linspace(0, 2 * np.pi, 1000)
y_tan = np.tan(x)

plt.plot(x, y_tan, label='tan(x)')
plt.ylim(-5, 5)

# Draw vertical asymptotes
for k in [0.5, 1.5]:
    plt.axvline(x=k * np.pi, color='red', linestyle='--', alpha=0.7)

plt.title('Tangent Function with Asymptotes')
plt.xlabel('x (radians)')
plt.ylabel('tan(x)')
plt.legend()
plt.grid(True)
plt.show()
````

## 6. Key Points to Remember

- **Sine** and **cosine** functions are smooth, wave-like, and periodic with a period of $2\pi $.
- **Tangent** has a period of $\pi$ and vertical asymptotes where it's undefined.
- Understanding these graphs helps in solving equations, modeling waves, and more.

## 7. Practice

Try plotting these functions with different amplitudes, periods, and phase shifts!

````python

A = 2      # Amplitude
omega = 0.5  # Frequency
phi = np.pi/4  # Phase shift

x = np.linspace(0, 4 * np.pi, 1000)
y = A * np.sin(omega * x + phi)
plt.plot(x, y, label='A*sin(ωx + φ)')
plt.title('General Sine Function')
plt.xlabel('x (radians)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
````
