# Trigonometry in Python: Polar Coordinates and Parametric Equations

Trigonometry is not just about right triangles—it's a powerful tool for describing points, curves, and motion in the 2D plane. Two important applications are **polar coordinates** and **parametric equations**. In this tutorial, we'll learn how to represent and plot points and functions using Python.


## 1. Polar Coordinates

### What Are Polar Coordinates?

Unlike Cartesian coordinates (x, y), **polar coordinates** use:

- `r`: the distance from the origin (0, 0)
- `θ` (theta): the angle from the positive x-axis (in radians or degrees)

A point in polar coordinates is written as **(r, θ)**.

### Converting Polar to Cartesian

To plot or compute with polar points, we often convert to Cartesian coordinates:

$$
x = r \cdot \cos(\theta)
$$
$$
y = r \cdot \sin(\theta)
$$

### Example: Convert and Plot a Polar Point

Let's convert the polar point (r=3, θ=π/4) to Cartesian and plot it.

````python
import math
import matplotlib.pyplot as plt

# Polar coordinates
r = 3
theta = math.pi / 4  # 45 degrees

# Convert to Cartesian
x = r * math.cos(theta)
y = r * math.sin(theta)

print(f"Cartesian coordinates: x={x:.2f}, y={y:.2f}")

# Plotting
plt.figure(figsize=(5,5))
plt.plot([0, x], [0, y], marker='o')
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.grid(True)
plt.title("Polar Point (r=3, θ=π/4)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

## 2. Polar Functions

A function in polar coordinates looks like `r = f(θ)`.

### Example: Plotting a Polar Function

Let's plot the polar equation of a circle: `r = 2`

````python
import numpy as np

theta = np.linspace(0, 2 * np.pi, 500)
r = 2 * np.ones_like(theta)

# Convert to Cartesian
x = r * np.cos(theta)
y = r * np.sin(theta)

plt.figure(figsize=(6,6))
plt.plot(x, y)
plt.gca().set_aspect('equal')
plt.title("Circle in Polar Coordinates (r = 2)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
````

## 3. Parametric Equations

In **parametric equations**, both `x` and `y` are defined as functions of a parameter, usually `t`:

$$
x = f(t)
$$
$$
y = g(t)
$$

### Example: Plotting a Circle Parametrically

Let's plot a circle of radius 2 centered at the origin.

````python
t = np.linspace(0, 2*np.pi, 500)
x = 2 * np.cos(t)
y = 2 * np.sin(t)

plt.figure(figsize=(6,6))
plt.plot(x, y)
plt.gca().set_aspect('equal')
plt.title("Circle Parametric Equations")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
````

## 4. Combining Polar and Parametric Forms

Many polar curves can be plotted using parametric equations by converting `r = f(θ)` into `x(θ) = r * cos(θ)`, `y(θ) = r * sin(θ)`.

### Example: Plotting a Spiral

$$
r = \theta
$$

````python
theta = np.linspace(0, 4 * np.pi, 1000)
r = theta

x = r * np.cos(theta)
y = r * np.sin(theta)

plt.figure(figsize=(6,6))
plt.plot(x, y)
plt.title("Archimedean Spiral (r = θ)")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.grid(True)
plt.show()
````

## Conclusion

- **Polar coordinates:** Useful for circular and rotational symmetry. Convert to (x, y) for plotting.
- **Parametric equations:** Great for describing complex curves and motion.
- Python (with `numpy` and `matplotlib`) makes it easy to explore and visualize both.
