# Trigonometry with Python: Periodic Functions and Harmonic Motion

Trigonometry is fundamental in understanding periodic phenomena—anything that repeats in cycles, such as sound waves, alternating current, or the motion of a pendulum. In this tutorial, we'll explore how to use Python to visualize and analyze periodic functions, and model harmonic motion, with practical examples in physics and engineering.

---

## 1. What are Periodic Functions?

A **periodic function** repeats its values in regular intervals or periods. The most common examples are the sine and cosine functions.

- **Sine Function:** $ y = \sin(x) $
- **Cosine Function:** $ y = \cos(x) $

The **period** is the smallest positive value $ P $ such that $ f(x) = f(x + P) $.

---

## 2. Visualizing Sine and Cosine with Python

Let's plot the sine and cosine functions using `numpy` and `matplotlib`.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of values from 0 to 2π
x = np.linspace(0, 2 * np.pi, 1000)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')
plt.title("Sine and Cosine Functions")
plt.xlabel("x (radians)")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

## 3. General Form of a Periodic Function

The general form for a sinusoidal function is:

$$
y(t) = A \sin(\omega t + \phi) + D
$$

Where:
- $ A $ = Amplitude (maximum displacement)
- $ \omega $ = Angular frequency ($ 2\pi f $), where $ f $ is frequency in Hz
- $ \phi $ = Phase shift (radians)
- $ D $ = Vertical shift

---

## 4. Simple Harmonic Motion (SHM)

**Simple Harmonic Motion** describes systems where the restoring force is proportional to displacement (like a mass on a spring or a pendulum for small angles).

The displacement as a function of time:

$$
x(t) = A \cos(\omega t + \phi)
$$

Let's simulate SHM for a mass on a spring:

```python
# Parameters
A = 1.0  # Amplitude (meters)
f = 0.5  # Frequency (Hz)
omega = 2 * np.pi * f  # Angular frequency
phi = 0  # Phase (radians)

t = np.linspace(0, 10, 1000)  # Time from 0 to 10 seconds

x = A * np.cos(omega * t + phi)

plt.figure(figsize=(8, 4))
plt.plot(t, x, label='x(t) = A cos(ωt + φ)')
plt.title("Simple Harmonic Motion: Mass on a Spring")
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.legend()
plt.grid(True)
plt.show()
```

## 5. Damped Harmonic Motion

In real systems, friction causes the amplitude to decrease over time:

$$
x(t) = A e^{-\gamma t} \cos(\omega t + \phi)
$$

Where $ \gamma $ is the damping coefficient.

```python
gamma = 0.1  # Damping coefficient

x_damped = A * np.exp(-gamma * t) * np.cos(omega * t + phi)

plt.figure(figsize=(8, 4))
plt.plot(t, x_damped, label='Damped SHM')
plt.title("Damped Harmonic Motion")
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.legend()
plt.grid(True)
plt.show()
```

## 6. Applications in Engineering and Physics

### a. **Alternating Current (AC) in Electronics**

The voltage in AC circuits varies sinusoidally:

$$
V(t) = V_0 \sin(\omega t)
$$

You can model and plot the voltage over time using the same approach.

### b. **Waves on a String or Sound Waves**

Waves can be described as:

$$
y(x, t) = A \sin(kx - \omega t)
$$

Where $ k $ is the wave number.

### c. **Pendulum Motion (Small Angles)**

For small oscillations, the angle θ follows:

$$
\theta(t) = \theta_0 \cos(\sqrt{\frac{g}{L}} t)
$$

Where $ g $ is gravitational acceleration and $ L $ is pendulum length.

## 7. Fourier Series: Combining Periodic Functions

Any periodic function can be represented as a sum of sines and cosines (Fourier series). This is fundamental in signal processing and vibration analysis.
