# Trigonometry with Python: A Beginner's Tutorial

This tutorial will walk you through the basics of working with trigonometric functions in Python, using the `math` library. We will cover the six fundamental trigonometric functions:

- Sine (`sin`)
- Cosine (`cos`)
- Tangent (`tan`)
- Secant (`sec`)
- Cosecant (`csc`)
- Cotangent (`cot`)

---

## 1. Prerequisites

You only need Python installed on your system. We'll use the built-in `math` module.

````python
import math
````

## 2. Angle Units: Degrees vs Radians

Python's `math` module works with angles in **radians** by default.  
To convert degrees to radians:

````python
angle_deg = 45
angle_rad = math.radians(angle_deg)
print(angle_rad)  # Output: 0.7853981633974483
````

## 3. Sine, Cosine, and Tangent

### Sine (`sin`)

The sine of an angle in radians:

````python
angle = math.radians(30)  # 30 degrees to radians
sine_value = math.sin(angle)
print(f"sin(30°) = {sine_value}")  # Output: 0.5
````

### Cosine (`cos`)

The cosine of an angle in radians:

````python
angle = math.radians(60)
cosine_value = math.cos(angle)
print(f"cos(60°) = {cosine_value}")  # Output: 0.5
````

### Tangent (`tan`)

The tangent of an angle in radians:

````python
angle = math.radians(45)
tangent_value = math.tan(angle)
print(f"tan(45°) = {tangent_value}")  # Output: 1.0
````

## 4. Secant, Cosecant, and Cotangent

Python’s `math` library does not provide these functions directly, but you can compute them as follows:

- **Secant**: `sec(x) = 1 / cos(x)`
- **Cosecant**: `csc(x) = 1 / sin(x)`
- **Cotangent**: `cot(x) = 1 / tan(x)`

Example:

````python
angle = math.radians(30)

secant_value = 1 / math.cos(angle)
cosecant_value = 1 / math.sin(angle)
cotangent_value = 1 / math.tan(angle)

print(f"sec(30°) = {secant_value}")
print(f"csc(30°) = {cosecant_value}")
print(f"cot(30°) = {cotangent_value}")
````

**Sample Output:**
```
sec(30°) = 1.1547005383792517
csc(30°) = 2.0
cot(30°) = 1.7320508075688774
```

## 6. Handling Special Cases

- Ensure input values do not make denominators zero (e.g., `sin(0)` for cosecant, `cos(90)` for secant).
- Use exception handling to manage division by zero:

````python
try:
    angle = math.radians(0)
    csc_value = 1 / math.sin(angle)
except ZeroDivisionError:
    csc_value = float('inf')  # or handle as needed

print(f"csc(0°) = {csc_value}")  # Output: inf
````

## 7. Visualization (Optional)

You can plot trigonometric functions using `matplotlib`:

````python

import numpy as np
import matplotlib.pyplot as plt

angles = np.linspace(0, 2 * np.pi, 360)
plt.plot(np.degrees(angles), np.sin(angles), label='sin(x)')
plt.plot(np.degrees(angles), np.cos(angles), label='cos(x)')
plt.plot(np.degrees(angles), np.tan(angles), label='tan(x)')
plt.xlim(0, 360)
plt.ylim(-2, 2)
plt.xlabel('Angle (degrees)')
plt.ylabel('Value')
plt.title('Trigonometric Functions')
plt.legend()
plt.grid(True)
plt.show()
````

## 8. Conclusion
