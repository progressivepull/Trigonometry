# Trigonometry in Python: Inverse Trigonometric Functions (arc sine, arc cosine, arc tangent)

This tutorial will guide you through using Python to work with the **inverse trigonometric functions**: **arc sine (`arcsin`)**, **arc cosine (`arccos`)**, and **arc tangent (`arctan`)**. You will learn their properties, how to compute them using Python, and practical applications.

## 1. Introduction to Inverse Trigonometric Functions

Inverse trigonometric functions allow you to find the angle when you know the ratio of sides in a right triangle.

- **arcsin(x)**: Returns the angle whose sine is `x`.
- **arccos(x)**: Returns the angle whose cosine is `x`.
- **arctan(x)**: Returns the angle whose tangent is `x`.

**Domain and Range:**

| Function | Domain          | Range                        |
|----------|----------------|------------------------------|
| arcsin   | [-1, 1]        | [-π/2, π/2]                  |
| arccos   | [-1, 1]        | [0, π]                       |
| arctan   | (-∞, ∞)        | (-π/2, π/2)                  |

## 2. Using Inverse Trigonometric Functions in Python

Python's `math` and `numpy` libraries provide these functions:

- `math.asin(x)`, `numpy.arcsin(x)`
- `math.acos(x)`, `numpy.arccos(x)`
- `math.atan(x)`, `numpy.arctan(x)`

Let's import them:

```python
import math
import numpy as np
```

### 2.1. Calculating arc sine, arc cosine, and arc tangent

```python
# Example value
x = 0.5

# Using math (single value)
asin_val = math.asin(x)     # radians
acos_val = math.acos(x)     # radians
atan_val = math.atan(x)     # radians

print("asin(0.5) =", asin_val)
print("acos(0.5) =", acos_val)
print("atan(0.5) =", atan_val)

# Converting radians to degrees
print("asin(0.5) in degrees =", math.degrees(asin_val))
```

### 2.2. Using numpy for arrays

```python
arr = np.array([0, 0.5, 1])

arcsin_arr = np.arcsin(arr)
arccos_arr = np.arccos(arr)
arctan_arr = np.arctan(arr)

print("arcsin:", np.degrees(arcsin_arr))
print("arccos:", np.degrees(arccos_arr))
print("arctan:", np.degrees(arctan_arr))
```

## 3. Properties of Inverse Trigonometric Functions

**1. arcsin(sin(x)) = x, for x in [-π/2, π/2]**

**2. arccos(cos(x)) = x, for x in [0, π]**

**3. arctan(tan(x)) = x, for x in (-π/2, π/2)**

**4. sin(arcsin(x)) = x, for x in [-1, 1]**

### Example in Python

```python
angle = math.radians(30)  # 30 degrees in radians
# Forward and inverse
sin_val = math.sin(angle)
angle_back = math.asin(sin_val)
print("Original angle (deg):", math.degrees(angle))
print("Recovered angle (deg):", math.degrees(angle_back))
```

## 4. Applications

### 4.1. Finding Angles in Triangles

Suppose you know the sides of a right triangle and want to find the angles:

```python
# Given sides
opposite = 3
adjacent = 4
hypotenuse = 5

# Find angle A (opposite/hypotenuse)
angle_A = math.degrees(math.asin(opposite / hypotenuse))
print("Angle A:", angle_A, "degrees")

# Find angle B (adjacent/hypotenuse)
angle_B = math.degrees(math.acos(adjacent / hypotenuse))
print("Angle B:", angle_B, "degrees")

# Find angle using arctan
angle_C = math.degrees(math.atan(opposite / adjacent))
print("Angle C:", angle_C, "degrees")
```

### 4.2. Real-World Example: Calculating the Angle of Elevation

Suppose a ladder 10 m long leans against a wall, reaching a height of 8 m. What's the angle the ladder makes with the ground?

```python
ladder = 10
height = 8

angle = math.degrees(math.asin(height / ladder))
print("Angle with ground:", angle, "degrees")
```

## 5. Visualization (Optional)

If you want to plot these functions:

```python
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100)
plt.plot(x, np.degrees(np.arcsin(x)), label="arcsin(x)")
plt.plot(x, np.degrees(np.arccos(x)), label="arccos(x)")
x_tan = np.linspace(-10, 10, 100)
plt.plot(x_tan, np.degrees(np.arctan(x_tan)), label="arctan(x)")
plt.xlabel("x")
plt.ylabel("Angle (degrees)")
plt.legend()
plt.title("Inverse Trigonometric Functions")
plt.show()
```

## 6. Summary

- Use `math.asin`, `math.acos`, `math.atan` for individual values; `numpy.arcsin`, `numpy.arccos`, `numpy.arctan` for arrays.
- Remember the domain and range of each inverse function.
- Inverse trigonometric functions are essential for calculating angles from side ratios in triangles and have many applications in engineering, physics, and geometry.
